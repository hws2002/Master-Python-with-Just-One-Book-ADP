# 1. row/column 선택

```python
#  선택한 값 변경하기
score = pd.DataFrame({'국어' : [95,75], '수학' : [75,90], '영어' : [90,95]}, index=['장화', '홍련'])
score
```

```python
score.loc[:,'영어'] = [100,100]
score
```

# 2. row/column 추가

```python
# (1) row 추가
new_students = pd.DataFrame({'국어' :[70, 65], '수학' : [65,100], '영어' : [95,-10]}, index = ['콩쥐', '팥쥐'])
score = pd.concat([score,new_students])
score
```

```python
new_students1 = pd.Series({'국어' : 85, '수학' : 55, '영어' : 95}, name = '해님')
score = pd.concat([score, new_students1.to_frame().T])
score
```

```python
# (2) column 추가
science = [80, 70, 90, 85, 75]
score['과학'] = science
score['학년'] = 1
score
```

```python
score['과학'] = score['과학'] + 5
score['총점'] = score['국어'] + score['수학'] + score['영어'] + score['과학']
score
```

# 3. row/column 삭제

```python
score_copy = score.copy()
```

```python
score_copy.drop('장화', inplace = True)
score_copy
```

```python
score_copy.drop(columns = ['영어', '과학', '학년', '총점'])
```

```python
score_copy
```

```python
# 행과 열을 동시에 지정하여 삭제하는 것을 불가능하다.
score_copy.drop('콩쥐',columns = '학년')
```

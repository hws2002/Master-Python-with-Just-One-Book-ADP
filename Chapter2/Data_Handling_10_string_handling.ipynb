```python
import pandas as pd

landmark = pd.DataFrame(
    {'name' : ['경복궁', '불국사', '첨성대', '광한루원', '창경궁'],
     'english name' : ['gyeongbokgung Palace',
                       'bulguksa',
                       'cheomseongdae observatory',
                       'gwanghanru',
                       'Changgyeonggung Palace'],
     'location' : ['서울특별시 종로구 사직로 161',
                   '경북 경주시 불국로 385',
                   '경북 경주시 인왕동 839-1',
                   '전북 남원시 요천로 1447',
                   '서울 종로구 창경궁로 185']
})
```

# 1. 문자열 대소문자 변환

```python
landmark
```

```python
landmark['name_upper'] = landmark['english name'].str.upper()
landmark['name_lower'] = landmark['english name'].str.lower()
landmark
```

```python
landmark['name_capitalize'] = landmark['english name'].str.capitalize()
landmark['name_title'] = landmark['english name'].str.title()
landmark[['english name', 'name_capitalize', 'name_title']]
```

# 2. 문자열 분리

```python
landmark['location'].str[3:6]
```

```python
series_split = landmark['location'].str.split()
series_split
```

```python
print('series_split의 타입 : ', type(series_split))
print('series_split.str.get(0) : \n',series_split.str.get(0))
print('series_split.str.get(0)의 타입 : ', type(series_split.str.get(0)))

print('series_split[0] : \n',series_split[0])
print('series_split[0]의 타입 : ',type(series_split[0]))
```

```python
# 띄어쓰기를 기준으로 분할,
landmark['location'].str.split(" ", expand = True)
```


# 3. 문자열 치환

```python
landmark['name_replace'] = landmark['english name'].str.replace(' ', '_')
landmark[['english name','name_replace']]
```

```python
# 'location' 컬럼에서 지역명 앞의 '서울특별시', '경북', '전북'을 각각 '서울', '경주', '남원' 으로 단순화하여 치환
landmark['simple_location'] = landmark['location'].str.replace("서울특별시", "서울").str.replace("경북 경주시", "경주").str.replace("전북 남원시", "남원")
landmark[['location', 'simple_location']]
```

```python
# 'location' 컬럼에서 '경주'라는 문자열만 추출해보자
landmark['address_with_경주'] = landmark['location'].str.extract('(경주)')

landmark[['location','address_with_경주']]
```

```python
# 만약 경주를 포함하는지 안하는지 알고싶다면
# 혹은 notnull()
landmark['address_with_경주'] = landmark['location'].str.extract('(경주)').notna()

landmark[['location','address_with_경주']]
```

```python
# landmark 데이터프레임의 location 컬럼에서 정규표현식을 활용하여 상세주소만 추출해보자
landmark['Detailed address'] = landmark['location'].str.extract('(\d+-?\d+)')

landmark[['name', 'location', 'Detailed address']]
```

# 4. 문자열 포함 여부 검사

```python
landmark['contains_py'] = landmark['english name'].str.contains('Palace')

landmark[['name_replace','contains_py']]
```

# 6. 문자열 제거

```python
import pandas as pd

landmark = pd.DataFrame(
    {'name' : [' 경복궁 ', '불국사', ' 첨성대', '광한루원 ', ' 창경궁'],
     'english name' : [' gyeongbokgung Palace ',
                       'bulguksa ',
                       'cheomseongdae observatory',
                       ' gwanghanru',
                       'Changgyeonggung Palace'],
     'location' : [' 서울특별시 종로구 사직로 161 ',
                   '경북 경주시 불국로 385',
                   '경북 경주시 인왕동 839-1 ',
                   ' 전북 남원시 요천로 1447',
                   '서울 종로구 창경궁로 185']
})
```

```python
landmark['name'] = landmark['name'].str.strip()
landmark['english name'] = landmark['english name'].str.strip()
landmark['location'] = landmark['location'].str.strip()
landmark
```

```python
# '경' 문자로 시작하는 경우 제거
landmark['name'] = landmark['name'].str.lstrip('경')

# 'e' 문자로 끝나는 경우 제거
landmark['english_name'] = landmark['english name'].str.rstrip('e')
landmark
```

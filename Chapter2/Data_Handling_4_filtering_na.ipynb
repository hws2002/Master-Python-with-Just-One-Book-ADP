# 1. 임의의 조건 탐색 및 데이터 수정

```python
import pandas as pd
students = pd.DataFrame({
    '이름' : ['장화', '홍련', '콩쥐', '팥쥐', '해님', '달님'],
    '국어' : [70, 85, None, 100, None, 85],
    '수학' : [65, 100, 80, 95, None, 70]
})
students
```

```python
students[students['이름'] == '장화']
# 그냥 dataframe형태로 변환
```

```python
students.loc[0,:]
# series형태로 변환
```

```python
students[ (students['국어']>=80) & (students['수학'] >= 80)]
```

```python
## loc[]에 조건문 사용
students.loc[6, '이름' : '수학'] = ['별님', 50, 60]
students
```

```python
students.loc[ (students['국어']>=80) & (students['수학'] >= 70), '합격'] = 'Pass'
students
```

```python
students.loc[students['합격'] != 'Pass','합격'] = 'Fail'
students
```

```python
# np.select() 활용
import numpy as np
condition_list = [(students['국어'] >= 90),
                  (students['국어'] >= 80) & (students['국어'] <= 90),
                  (students['국어'] >= 70) & (students['국어'] < 80)]
choice_list = ['A', 'B', 'C']
students['점수'] = np.select(condition_list, choice_list, default = 'F')
students
```

```python
# 결측값 대체
health = pd.DataFrame(
    { '연도' : [2017, 2018, 2019, 2020, 2021, 2022],
      '키' : [160, 162, 165, None, None, 166],
      '몸무게' : [53, 52, None, 50, 51, 54],
      '시력' : [1.2, None, 1.2, 1.2, 1.1, 0.8],
      '병결' : [None, None, None, 2, None, 1]

    }
)
health
```

```python
health.fillna(0)
```

```python
health.mean() #열 평균
```

```python
health.fillna(health.mean())
```

```python
# 병결 컬럼의 결측치를 0으로, 몸무게 컬럼의 결측치를 평균으로 대체해보자
# print(health)
health['병결'] = health['병결'].fillna(0)
health['몸무게'] = health['몸무게'].fillna(health['몸무게'].mean())
```

```python
health
```

```python
# 키와 시력 컬럼을 결측값이 등장하기 바로 직전 값으로 대체해보자
health.fillna(method = 'pad', inplace = True)
health
```

# 3. 중복행 삭제

```python
# health 데이터세트에서 '키' 컬럼이 가지고 있는 값에는 어떤 것이 있는지 중복 없이 확인해보자
health['키'].drop_duplicates()
```

```python
# 이 경우에는 파이썬의 기본 함수인 set()을 대신 사용할 수도 있다.
# duplicates() 함수의 출력 결과는 pd.Series형식으로, set()함수의 출력 결과는 tuple 형식이다
set(health['키'])
```

```python
# 여러 개의 컬럼을 가진 데이터에 대해 drop_duplicates()함수를 적용하면 각 행을 하나의 객체로 보고 중복을 제거한다
health[['시력','병결']].drop_duplicates()
```

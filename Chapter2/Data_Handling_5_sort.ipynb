```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
iris = pd.DataFrame(iris.data, columns = iris.feature_names)
iris.head(3)
```

```python
# 인덱스 기준 정렬
iris.sort_index(axis = 0, ascending = False, inplace = True)
iris.head()
```

```python
iris.sort_index(axis = 1, ascending = True, inplace = False)
```

```python
iris.sort_values(by = ['petal width (cm)', 'sepal length (cm)'])
```

```python
iris.sort_index(ascending = True, inplace = True)
iris
```

```python
# 0번째 행만 보고, 0번째 행의 값들이 증가하는 모습이 되도록 컬럼들을 조정함
iris.sort_values(by=0, axis = 1)
```

```python
# index 변경
iris.index = list('row'+str(i) for i in range(0,150))
```

```python
iris.index
```

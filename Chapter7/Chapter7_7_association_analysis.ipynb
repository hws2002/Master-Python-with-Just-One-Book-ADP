# 1. Run-Test

```python
import pandas as pd
data = ['a','a','b','b','a','a','a','a','b','b','b','b','b','a','a','b','b','a','b','b']
test_df = pd.DataFrame(data, columns = ['product'])
test_df
```

```python
from statsmodels.sandbox.stats.runs import runstest_1samp

# Run-test를 위한 데이터 변환
test_df.loc[test_df['product'] == 'a', 'product'] = 1
test_df.loc[test_df['product'] == 'b', 'product'] = 0

test_df
```

```python
runstest_1samp(test_df['product'], cutoff=0.5, correction=True)
```

p-value가 0.05보다 크므로 귀무가설을 기각할 수 없다.  
즉, 해당 데이터의 연속적인 관측값은 임의적이라고 볼 수 있다.(구매에 연관이 없다)  


# 2. 연관규칙분석


```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
dataset = [['Apple', 'Beer' ,'Rice', 'Chicken'],
           ['Apple', 'Beer' ,'Rice'],
           ['Apple', 'Beer'],
           ['Apple', 'Bananas'],
           ['Milk', 'Beer' ,'Rice', 'Chicken'],
           ['Milk', 'Beer' ,'Rice'],
           ['Milk', 'Beer'],
           ['Apple', 'Bananas']]
te = TransactionEncoder()
te_ary = te.fit_transform(dataset)
print(te.columns_)
te_ary
```

```python
df = pd.DataFrame(te_ary, columns = te.columns_)
df
```

```python
# 최소 지지돠 0.6 이상인 품목만 추출해보자
from mlxtend.frequent_patterns import apriori
apriori(df, min_support = 0.6, use_colnames = True)
```

```python
frequent_itemsets = apriori(df, min_support = 0.3, use_colnames = True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x : len(x))
frequent_itemsets
```

grocerires 데이터세트로 연관규칙분석을 실행해보자.  

```python
df = pd.read_csv('https://raw.githubusercontent.com/ADPclass/ADP_book_ver01/main/data/groceries.csv')
df
```

```python
print(df.iloc[:,0])
```

```python
df_split = df.iloc[:,0].str.split(',', expand = True)
df_split_ary = df_split.values

groceries = []
for i in range(len(df_split_ary)):
  temp = list(filter(None, df_split_ary[i]))
  groceries.append(temp)

groceries[:3]
```

```python
groceries_tr = te.fit_transform(groceries)
print(te.columns_)
groceries_tr= pd.DataFrame(groceries_tr, columns = te.columns_)
groceries_tr
```

```python
groceries_ap = apriori(groceries_tr,min_support = 0.01, use_colnames = True)
groceries_ap
```

```python
from mlxtend.frequent_patterns import association_rules

groceries_ap['num_itemsets'] = groceries_ap['itemsets'].apply(lambda x : len(x))
groceries_ap
```

```python
rules = association_rules(groceries_ap, metric="confidence", min_threshold=0.3, num_itemsets = len(groceries_tr))
rules
```

```python
# 규칙의 길이가 2 이상, 신뢰도가 0.4이상, 향상도가 3이상인 규칙만 추출해보자
rules['antecedent_len'] = rules['antecedents'].apply(lambda x : len(x))
rules[(rules['antecedent_len'] >= 2) &
      (rules['confidence'] >= 0.4) &
      (rules['lift'] >= 3)]
```

```python
rules[(len(rules['antecedents']) >= 2) &
      (rules['confidence'] >= 0.4) &
      (rules['lift'] >= 3)]
```
